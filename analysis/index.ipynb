# Demos for mix-SQP paper

These webpages contain demos of **mix-SQP**, a fast algorithm for maximum-likelihood estimation of mixture proportions, and examples comparing mix-SQP against other alternative approaches, such as EM and interior-point methods.

The webpages were generated by rendering the [Jupyter notebooks](http://jupyter.org) into webpages using [nbconvert](http://nbconvert.readthedocs.io).

For more information, including Setup Instructions, please see the README in the [git repository](https://github.com/stephenslab/mixopt-paper).

These are the demonstrations we have implemented:

1. [Illustration of mix-SQP solver applied to a small data set, and a large one](mixSQP_example.html).

2. [mix-SQP demo with different low-rank matrix approximations](low_rank_approximation.html).

3. [Comparing performance and accuracy of EM, IP and mix-SQP algorithms](mixSQP_REBayes_EM.html).

4. [Profiling adaptive shrinkage computations with mix-SQP and IP solvers](ash_computation_time.html).
