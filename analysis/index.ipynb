# Numerical experiments for mix-SQP paper

These webpages contain demonstrations of **mix-SQP**, a fast algorithm for computing maximum-likelihood estimates of mixture proportions, and code and results for numerical experiments comparing mix-SQP against other optimization appraoches.

All the webpages were generated by rendering the [Jupyter notebooks](http://jupyter.org) into webpages using [nbconvert](http://nbconvert.readthedocs.io). The results should be reproducible on your computer. See the [README](https://github.com/stephenslab/mixopt-paper) for details.

These are the demonstrations and experiments we have implemented.

1. [Illustration of mix-SQP solver applied to a small data set, and a large one](mixSQP_example.html).
